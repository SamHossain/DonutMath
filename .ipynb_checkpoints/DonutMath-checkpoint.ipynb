A few preliminaries before we start any code blocks. First is a basic understanding of polar coordinates. Basically, a conversion from the XY coordinate plane to one that uses an angle, $\theta$ and a radius, r. (e.g. $x = r \cdot cos(\theta)$) Second is an understanding of <a href="https://en.wikipedia.org/wiki/Rotation_matrix">rotation matricies. </a> It is easier to look at a 2D rotation and then extend the notion. In the standard setting, the matrix is $$
R = \begin{bmatrix}
cos(\theta) & -sin(\theta) \\
sin(\theta) & cos(\theta) \\
\end{bmatrix}
$$
This is an extension of polar coordinates. A proof of this is <a href="">here.</a> We can easily extend this to three dimentions. Note that adding a z axis to this rotation does not effect a rotation about the origin. So, 
$$
\begin{bmatrix}
cos(\theta) & -sin(\theta) \\
sin(\theta) & cos(\theta) \\
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
\end{bmatrix}
$$
and when we add z, we see that under $R_z$ z is unchanged:
$$
R_z = \begin{bmatrix}
cos(\theta) & -sin(\theta) & 0\\
sin(\theta) & cos(\theta) & 0\\
0 & 0 & 1\\
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
z \\
\end{bmatrix}
$$

```python
#few imports
%matplotlib inLine
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import math
```

So, for the first part we will generate our taurus. Lets generate the first circle with a loop that has radius 2. And offset it by 3 units. theta_iter controls how few or many points compose our circle. When we set theta, we define it to be some fraction of $2\pi$, indicating how far along the circle we are.

```python
theta_iter = 20
r = 3
d_r = 2
#plotting purposes
x_data = []
y_data = []
z_data = []
for i in range(theta_iter):
    theta = 2*math.pi*i/theta_iter
    x, y, z = d_r*math.cos(theta), d_r*math.sin(theta), 0
    x = x + r
    
    #plotting purposes
    x_data.append(x)
    y_data.append(y)
    z_data.append(z)

#plotting purposes
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(x_data, y_data, z_data)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

Now, that we have a circle, we can rotate it about the y-axis in order to form a taurus. Here, instead of the z-axis, we use the matrix:
$$R_y = \begin{bmatrix}
cos(\theta) & 0 &-sin(\theta)\\
0 & 1 & 0 \\
cos(\theta) & 0 &sin(\theta)\\
\end{bmatrix}
$$
Similar to the above, we take the circle and loop it over the $2\pi$ in phi_iter increments.

```python
phi_iter = 20
for j in range(phi_iter):
    #x,y,z are the coordinates of our circle
    x_, y_, z_ = x, y, z
    phi = 2*math.pi*j/phi_iter
    x_, y_, z_ = x*math.cos(phi) + z*math.sin(phi), y, -1* x*math.sin(phi) - z*math.cos(phi)
```

Combining this loop with the one to generate our circle, we have:

```python
theta_iter = 35
r = 4
d_r = 2
#plotting purposes
x_data = []
y_data = []
z_data = []
for i in range(theta_iter):
    theta = 2*math.pi*i/theta_iter
    x, y, z = d_r*math.cos(theta), d_r*math.sin(theta), 0
    x = x + r
    
    for j in range(phi_iter):
        #x,y,z are the coordinates of our circle
        x_, y_, z_ = x, y, z
        phi = 2*math.pi*j/phi_iter
        x_, y_, z_ = x*math.cos(phi) + z*math.sin(phi), y, -1* x*math.sin(phi) - z*math.cos(phi)
        #plotting purposes
        x_data.append(x_)
        y_data.append(y_)
        z_data.append(z_)

#plotting purposes
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(x_data, y_data, z_data)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

Now that we have a basic taurus, we can apply multiple rotation matricies to it in order to produce the effect of flying through the air. By applying $R_z, R_y, and R_z$ to our donut one at a time is the same as multiplying them together, which we show below. Don't worry about how messy it looks, just remember it by its simpler three matrix form. This is only useful to speed up coding the rotation.$$
\begin{bmatrix}
cos(\alpha)cos(\beta) & cos(\alpha)sin(\beta)sin(\gamma) - sin(\alpha)cos(\gamma) &cos(\alpha)sin(\beta)cos(\gamma) + sin(\alpha)sin(\gamma)\\
sin(\alpha)cos(\beta) & sin(\alpha)sin(\beta)sin(\gamma) + cos(\alpha)cos(\gamma) &sin(\alpha)sin(\beta)cos(\gamma) - cos(\alpha)sin(\gamma)\\
-sin(\beta) & cos(\beta)sin(\gamma) & cos(\beta)cos(\gamma)\\
\end{bmatrix}
$$
Here, alpha, beta, and gamma control how the three tilt directions move. We will increase beta and gamma to slowly revolve the taurus over the x and y axes.

```python
#shortcuts for defining a matrix
def sinsin(a, b):
    return math.sin(a)*math.sin(b)

def coscos(a, b):
    return math.cos(a)*math.cos(b)

def sincos(a, b):
    return math.sin(a)*math.cos(b)

def sinsinsin(a,b,c):
    return math.sin(a)*math.sin(b)*math.sin(c)

def sinsincos(a,b,c):
    return math.sin(a)*math.sin(b)*math.cos(c)

def sincoscos(a,b,c):
    return math.sin(a)*math.cos(b)*math.cos(c)
```

```python
alpha, beta, gamma = 0, 0, 0
#how fast we want to be rotating
pitch = .001
roll = .002
#we want a loop to run indefinitely
while True:
    beta += pitch
    gamma += roll
    x_, y_, z_ = 0, 0, 0 #These should actually be the coordinates of a point on the taurus
    
    #rotation matrix applied to x_, y_, z_
    X = x_*coscos(alpha, beta) + y_*(sinsincos(gamma, beta, alpha) - sincos(alpha, gamma)) + z_*(sincoscos(beta, alpha, gamma) + sinsin(alpha, gamma))
    Y = x_*sincos(alpha, beta) + y_*(sinsinsin(alpha, beta, gamma) + coscos(alpha, gamma)) + z_*(sinsincos(alpha, beta, gamma) - sincos(gamma, alpha))
    Z = -1*x_*math.sin(beta) + y_*sincos(gamma, beta) + z_*coscos(beta, gamma)
```

Next, we can add the lighting effects onto the donut. The light will point from some arbitrary direction towards the donut. This can be represented as a a 3D vector. Then we can see how directly the light source is focused on that surface by finding the surfaces normal vector and seeing how closely it aligns with the lighting direction. 
$$
\overrightarrow{light source}
\;
\overrightarrow{towards}
\;
\overleftarrow{away}
$$
So we need a way numerically identify when something is facing towards the lightsource versus away from the source. This can be achieved by using the formula: $$
y = cos(\theta) = \frac{u \cdot v}{|u||v|}
$$
y will be a number between -1 (away) and 1 (towards) the lightsource. What this is actually calculating is the angle, $\theta$ between two vectors, $u$ and $v$. If the angle is close to 0 (or 360) the two arrow are close to parallel and $cos(0) = 1$.
$$
$$
So now we have a way to quantify how 'close' we are to a light source but we still need to figure out the normal vector of our taurus. There is a kind of difficult way to compute this, using the formula of a taurus to calculate its <a href="https://en.wikipedia.org/wiki/Normal_(geometry)">gradient</a> at a point. But a shortcut is to use some of the nice geometric properties that a taurus has. If we look at any cross section of the taurus, we see a circle. Taking the center of the circle and drawing a line through any point constructs a vector. This vector will be the normal at that point. This is easier said in pictures.
